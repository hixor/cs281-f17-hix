```python
import numpy as np
import pandas as pd
import numpy.linalg as la
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import linear_model
```

```python
data = np.loadtxt("CASP.csv", delimiter = ",", skiprows = 1)
```

```python
y = data[:,0]
X = data[:,1:] 
norm_X = (X-np.mean(X)/np.std(X))
b_X = np.zeros((norm_X.shape[0],norm_X.shape[1]+1))
b_X[:,:-1] = norm_X
```

```python
n = len(y)
m = b_X.shape[1]

train_y = y[:n*90]
test_y = y[n*90:]

train_X = b_X[:n*90,:]
test_X = b_X[n*90:,:]
```

```python
sigma = 1.0
t = 10
Sig_inv = t*np.eye(m)
```

```python
L = la.cholesky(Sig_inv)
```

```python
train_X_tilde = np.vstack((train_X/sigma,L))
train_y_tilde = np.append(train_y/sigma, np.zeros(m))
```

```python
Q, R = la.qr(train_X_tilde)
```

```python
p = np.dot(Q.T,train_y_tilde)
w_hat = np.dot(la.inv(R),p)
```

```python
w_hat.shape
```

```python
np.sqrt(np.sum((np.dot(train_X,w_hat)-train_y)**2))
```
