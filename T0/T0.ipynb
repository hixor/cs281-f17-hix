```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Question 6, Gradient Check

def f(x):
    return np.cos(x) + x**2 + np.exp(x)

def grad_f(x):
    return -np.sin(x) + 2*x + np.exp(x)

def grad_check(x, epsilon):
    return (f(x + epsilon) - f(x - epsilon))/(2.0*epsilon)
```

```python
x_list = np.arange(0,1,0.001)
```

```python
analytic = [grad_f(x) for x in x_list]
epsilon_1 = [grad_check(x,1) for x in x_list]
epsilon_05 = [grad_check(x,0.5) for x in x_list]
epsilon_01 = [grad_check(x,0.1) for x in x_list]
epsilon_001 = [grad_check(x,0.01) for x in x_list]
epsilon_0001 = [grad_check(x,0.001) for x in x_list]
```

```python
plt.figure(figsize=(20,10))

plt.plot(range(len(analytic)),analytic,'green')
plt.plot(range(len(epsilon_1)),epsilon_1,'blue')
plt.plot(range(len(epsilon_05)),epsilon_05,'red')
plt.plot(range(len(epsilon_01)),epsilon_01,'red')
plt.plot(range(len(epsilon_001)),epsilon_001,'orange')
plt.plot(range(len(epsilon_0001)),epsilon_0001,'purple')
plt.show()
```
